```python
# Liste qui va contenir tous les étudiants
classe = []

# Demander le nombre d'étudiants
nb_etudiants = int(input("Entrez le nombre d'étudiants : "))

# Boucle pour chaque étudiant
for i in range(nb_etudiants):
    print(f"\n--- Étudiant {i + 1} ---")
    
    nom = input("Nom de l'étudiant : ")
    
    notes = []
    for j in range(3):
        note = float(input(f"Entrez la note {j + 1} : "))
        notes.append(note)
    
    moyenne = sum(notes) / len(notes)
    
    # Dictionnaire étudiant
    etudiant = {
        "nom": nom,
        "notes": notes,
        "moyenne": moyenne
    }
    
    # Ajouter l'étudiant à la classe
    classe.append(etudiant)

# Analyse globale
somme_moyennes = 0

print("\n===== RAPPORT DE LA CLASSE =====")
for etu in classe:
    print(f"\nÉtudiant : {etu['nom']}")
    print(f"Notes : {etu['notes']}")
    print(f"Moyenne : {etu['moyenne']:.2f}")
    
    somme_moyennes += etu["moyenne"]

moyenne_generale = somme_moyennes / nb_etudiants
print(f"\n📊 Moyenne générale de la classe : {moyenne_generale:.2f}")
```

```python
classe = []

nb_etudiants = int(input("Entrez le nombre d'étudiants : "))

for i in range(nb_etudiants):
    print(f"\n--- Étudiant {i + 1} ---")
    
    nom = input("Nom de l'étudiant : ")
    notes = []

    for j in range(3):
        note = float(input(f"Entrez la note {j + 1} : "))
        notes.append(note)

    moyenne = sum(notes) / len(notes)

    # Mention et statut
    if moyenne >= 90:
        mention = "Excellent"
    elif moyenne >= 70:
        mention = "Très bien"
    elif moyenne >= 50:
        mention = "Réussi"
    else:
        mention = "Échec"

    statut = "Admis" if moyenne >= 50 else "Ajourné"

    etudiant = {
        "nom": nom,
        "notes": notes,
        "moyenne": moyenne,
        "mention": mention,
        "statut": statut
    }

    classe.append(etudiant)

# Classement par moyenne décroissante
classe = sorted(classe, key=lambda x: x["moyenne"], reverse=True)

# Analyse globale
toutes_les_notes = []
for etu in classe:
    toutes_les_notes.extend(etu["notes"])

meilleure_note = max(toutes_les_notes)
plus_faible_note = min(toutes_les_notes)

print("\n===== 📊 RAPPORT FINAL DE LA CLASSE =====")

for i, etu in enumerate(classe, start=1):
    print(f"\n🏅 Rang {i}")
    print(f"Nom : {etu['nom']}")
    print(f"Notes : {etu['notes']}")
    print(f"Moyenne : {etu['moyenne']:.2f}")
    print(f"Mention : {etu['mention']}")
    print(f"Statut : {etu['statut']}")

moyenne_generale = sum([e["moyenne"] for e in classe]) / nb_etudiants

print("\n===== 📈 STATISTIQUES GÉNÉRALES =====")
print(f"Moyenne générale : {moyenne_generale:.2f}")
print(f"Meilleure note : {meilleure_note}")
print(f"Plus faible note : {plus_faible_note}")
```

```python
def saisir_notes():
    notes = []
    for i in range(3):
        note = float(input(f"Entrez la note {i + 1} : "))
        notes.append(note)
    return notes


def calculer_moyenne(notes):
    return sum(notes) / len(notes)


def donner_mention(moyenne):
    if moyenne >= 90:
        return "Excellent"
    elif moyenne >= 70:
        return "Très bien"
    elif moyenne >= 50:
        return "Réussi"
    else:
        return "Échec"


def statut_etudiant(moyenne):
    return "Admis" if moyenne >= 50 else "Ajourné"


def saisir_etudiant():
    nom = input("Nom de l'étudiant : ")
    notes = saisir_notes()
    moyenne = calculer_moyenne(notes)

    return {
        "nom": nom,
        "notes": notes,
        "moyenne": moyenne,
        "mention": donner_mention(moyenne),
        "statut": statut_etudiant(moyenne)
    }


def afficher_rapport(classe):
    print("\n===== 📊 RAPPORT FINAL DE LA CLASSE =====")

    classe = sorted(classe, key=lambda x: x["moyenne"], reverse=True)

    for i, etu in enumerate(classe, start=1):
        print(f"\n🏅 Rang {i}")
        print(f"Nom : {etu['nom']}")
        print(f"Notes : {etu['notes']}")
        print(f"Moyenne : {etu['moyenne']:.2f}")
        print(f"Mention : {etu['mention']}")
        print(f"Statut : {etu['statut']}")

    moyenne_generale = sum(e["moyenne"] for e in classe) / len(classe)
    print(f"\n📈 Moyenne générale de la classe : {moyenne_generale:.2f}")


# === Programme principal ===
classe = []
nb_etudiants = int(input("Entrez le nombre d'étudiants : "))

for _ in range(nb_etudiants):
    classe.append(saisir_etudiant())

afficher_rapport(classe)
```
