```python
import pandas as pd

# Création des données
data = {
    "nom": ["Paul", "Sarah", "Luc", "Alice", "Jean", "Marie", "David", "Emma", "Tom", "Chloe"],
    "age": [21, 23, 20, 22, 21, 23, 22, 20, 21, 22],
    "note": [65, 88, 72, 90, 72, 85, 60, 78, 55, 95]
}

# Création du DataFrame
df = pd.DataFrame(data)

# Sauvegarde dans un fichier CSV
df.to_csv("etudiants.csv", index=False)

print("Fichier 'etudiants.csv' généré avec succès !")
```

```python
import pandas as pd
import numpy as np

# Création des données "sales"
data = {
    "nom": ["Paul", "Sarah", "Luc", "", "Alice", "Jean", "Luc", "Tom", "Chloe", "Marie", "Emma", "David", "Sarah"],
    "age": [21, 23, 20, 22, np.nan, 21, 20, 21, 22, 23, np.nan, 22, 23],
    "note": [65, 88, 72, 90, 85, 72, 72, 55, 95, 85, 78, 60, 88]
}

# Création du DataFrame
df_sale = pd.DataFrame(data)

# Sauvegarde dans un fichier CSV
df_sale.to_csv("etudiants_sale.csv", index=False)

print("Fichier 'etudiants_sale.csv' généré avec succès !")
```

```python
import pandas as pd

# Charger le fichier CSV
df = pd.read_csv("etudiants_sale.csv")

# 🔹 Vérification des valeurs manquantes
print("Valeurs manquantes par colonne :")
print(df.isnull().sum())

# 🔹 Supprimer les lignes avec valeurs manquantes
df_clean = df.dropna()

# 🔹 Supprimer les doublons
df_clean = df_clean.drop_duplicates()

# 🔹 Vérifier les types de données
print("\nTypes de données avant conversion :")
print(df_clean.dtypes)

# 🔹 Conversion des types si nécessaire
df_clean["age"] = df_clean["age"].astype(int)
df_clean["note"] = df_clean["note"].astype(float)

# 🔹 Afficher le dataset nettoyé
print("\nDataset nettoyé :")
print(df_clean)
```

```python
import pandas as pd
import numpy as np

# Charger le fichier CSV "sale"
df = pd.read_csv("etudiants_sale.csv")

print("=== Dataset original ===")
print(df)

# 1️⃣ Nettoyage
df_clean = df.dropna()          # Supprimer les lignes avec NaN
df_clean = df_clean.drop_duplicates()  # Supprimer doublons

# S'assurer que les types sont corrects
df_clean["age"] = df_clean["age"].astype(int)
df_clean["note"] = df_clean["note"].astype(float)

print("\n=== Dataset nettoyé ===")
print(df_clean)

# 2️⃣ Statistiques descriptives
moyenne_generale = df_clean["note"].mean()
note_max = df_clean["note"].max()
note_min = df_clean["note"].min()
mediane = df_clean["note"].median()
variance = df_clean["note"].var()
ecart_type = df_clean["note"].std()

# 3️⃣ Rapport final
print("\n===== RAPPORT DES NOTES =====")
print(f"Nombre d'étudiants : {len(df_clean)}")
print(f"Moyenne générale : {moyenne_generale:.2f}")
print(f"Note maximale : {note_max}")
print(f"Note minimale : {note_min}")
print(f"Médiane : {mediane}")
print(f"Variance : {variance:.2f}")
print(f"Ecart-type : {ecart_type:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1️⃣ Charger le dataset "sale"
df = pd.read_csv("etudiants_sale.csv")

# 2️⃣ Nettoyage
df_clean = df.dropna()          # Supprimer lignes avec valeurs manquantes
df_clean = df_clean.drop_duplicates()  # Supprimer doublons
df_clean["age"] = df_clean["age"].astype(int)
df_clean["note"] = df_clean["note"].astype(float)

# 3️⃣ Statistiques descriptives
moyenne_generale = df_clean["note"].mean()
note_max = df_clean["note"].max()
note_min = df_clean["note"].min()
mediane = df_clean["note"].median()
variance = df_clean["note"].var()
ecart_type = df_clean["note"].std()

# 4️⃣ Affichage du rapport
print("===== RAPPORT DES NOTES DES ÉTUDIANTS =====")
print(f"Nombre d'étudiants : {len(df_clean)}")
print(f"Moyenne générale : {moyenne_generale:.2f}")
print(f"Note maximale : {note_max}")
print(f"Note minimale : {note_min}")
print(f"Médiane : {mediane}")
print(f"Variance : {variance:.2f}")
print(f"Ecart-type : {ecart_type:.2f}")

# 5️⃣ Visualisation simple
plt.figure(figsize=(10,6))
plt.bar(df_clean["nom"], df_clean["note"], color='skyblue')
plt.axhline(moyenne_generale, color='red', linestyle='--', label=f"Moyenne : {moyenne_generale:.2f}")
plt.title("Notes des étudiants")
plt.xlabel("Nom")
plt.ylabel("Note")
plt.ylim(0, 100)
plt.legend()
plt.show()
```
