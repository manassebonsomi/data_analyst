```python
import pandas as pd

# Charger le dataset
df = pd.read_csv("etudiants.csv")

# 1️⃣ Moyenne des notes par âge
moyenne_par_age = df.groupby("age")["note"].mean()
print("Moyenne des notes par âge :")
print(moyenne_par_age)

# 2️⃣ Note maximale par âge
max_par_age = df.groupby("age")["note"].max()
print("\nNote maximale par âge :")
print(max_par_age)

# 3️⃣ Création de la colonne 'performance'
def performance(note):
    if note >= 85:
        return "Excellent"
    elif note >= 70:
        return "Bien"
    else:
        return "Faible"

df["performance"] = df["note"].apply(performance)

# 4️⃣ Nombre d'étudiants par performance
nb_parformance = df.groupby("performance")["nom"].count()
print("\nNombre d'étudiants par performance :")
print(nb_parformance)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("etudiants.csv")

# Créons aussi la colonne 'performance'
df["performance"] = df["note"].apply(lambda n: "Excellent" if n >= 85 else "Bien" if n >= 70 else "Faible")
```

```python
sns.histplot(df["note"], bins=5, kde=True, color="skyblue")
plt.title("Distribution des notes")
plt.xlabel("Note")
plt.ylabel("Nombre d'étudiants")
plt.show()
```

```python
sns.boxplot(x="performance", y="note", data=df, palette="Set2")
plt.title("Notes des étudiants par performance")
plt.show()
```

```python
sns.lmplot(x="age", y="note", data=df, hue="performance", height=6)
plt.title("Relation âge / note selon la performance")
plt.show()
```

```python
corr = df[["age", "note"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Charger le dataset "sale"
df = pd.read_csv("etudiants_sale.csv")

# 2️⃣ Nettoyage
df_clean = df.dropna()          # Supprimer NaN
df_clean = df_clean.drop_duplicates()  # Supprimer doublons
df_clean["age"] = df_clean["age"].astype(int)
df_clean["note"] = df_clean["note"].astype(float)

# 3️⃣ Ajouter la colonne 'performance'
df_clean["performance"] = df_clean["note"].apply(
    lambda n: "Excellent" if n >= 85 else "Bien" if n >= 70 else "Faible"
)

# 4️⃣ Statistiques descriptives
moyenne = df_clean["note"].mean()
note_max = df_clean["note"].max()
note_min = df_clean["note"].min()
mediane = df_clean["note"].median()
variance = df_clean["note"].var()
ecart_type = df_clean["note"].std()

print("===== RAPPORT DES NOTES =====")
print(f"Nombre d'étudiants : {len(df_clean)}")
print(f"Moyenne générale : {moyenne:.2f}")
print(f"Note maximale : {note_max}")
print(f"Note minimale : {note_min}")
print(f"Médiane : {mediane}")
print(f"Variance : {variance:.2f}")
print(f"Ecart-type : {ecart_type:.2f}\n")

# 5️⃣ Corrélation
correlation = df_clean["age"].corr(df_clean["note"])
print(f"Corrélation âge / note : {correlation:.2f}\n")

# 6️⃣ Visualisations avancées

# Histogramme des notes
sns.histplot(df_clean["note"], bins=5, kde=True, color="skyblue")
plt.title("Distribution des notes")
plt.xlabel("Note")
plt.ylabel("Nombre d'étudiants")
plt.show()

# Boxplot par performance
sns.boxplot(x="performance", y="note", data=df_clean, palette="Set2")
plt.title("Notes des étudiants par performance")
plt.show()

# Scatterplot âge vs note avec regression
sns.lmplot(x="age", y="note", data=df_clean, hue="performance", height=6)
plt.title("Relation âge / note selon la performance")
plt.show()

# Heatmap corrélation
corr_matrix = df_clean[["age", "note"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
```
